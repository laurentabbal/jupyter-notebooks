**Programme complet**

```
# Bilan de matière de l'équation : 2H2 + O2 -> 2H2O
print("Bonjour, nous allons étudier la réaction entre le dihydrogène H2 et le dioxygène O2")
V1 = float(input("Entrez la valeur de la quantité de matière de dihydrogène en mL : "))
V2 = float(input("Entrez la valeur de la quantité de matière de dioxygène en mL : "))

# ligne à ajouter pour tranformer les volumes en mL en quantité de matière
Vm = 24 # en L/mol - c'est le volume molaire des gaz à 20°C sous 1 atm
n1 = V1 / (1000 * Vm)
n2 = V2 / (1000 * Vm)

# quantité de matière du produit à l'état initial
n3 = 0

# hypothèse 1 : H2 est le réactif limitant
# n1 - 2*xmax1 = 0
xmax1 = n1 / 2

# hypothèse 2 : 02 est le réactif limitant
# n2 - xmax2 = 0
xmax2 = n2

if xmax1 < xmax2:
    xf = xmax1
    print("Le dihydrogène est le réactif limitant.")

elif xmax2 < xmax1:
    xf = xmax2
    print('Le dioxygène est le réactif limitant.')
    
elif xmax1 == xmax2:
    xf = xmax1
    print("Les réactifs ont été introduits dans des proportions stoechiométriques.")
    
# quantités de matière à l'état final   
n1f = n1 - 2*xf    
n2f = n2 - xf  
n3f = n3 + 2*xf   
    
# affichage de l'état du système à l'état final
print("A l'état final, la quantité de matière de dihydrogène vaut : ", n1f, " mol")
print("A l'état final, la quantité de matière de dioxygène vaut : ", n2f, " mol")
print("A l'état final, la quantité de matière d'eau vaut : ", n3f, " mol")
```

### Question 3 : réaction de combustion du méthane

$CH_{4~(g)} + 20_{2~(g)} \rightarrow CO_{2~(g)} + 2H_{2}O_{~(l)}$

```
# Bilan de matière de l'équation : CH4 + 2O2 -> CO2 + 2H2O
print("Bonjour, nous allons étudier la réaction entre le méthane CH4 et le dioxygene O2")
n1 = float(input("Entrez la valeur de la quantité de matière de méthane en mol : "))
n2 = float(input("Entrez la valeur de la quantité de matière de dioxygène en mol : "))
           
# quantité de matière des produits à l'état initial
n3 = 0
n4 = 0

# hypothèse 1 : CH4 est le réactif limitant
# n1 - xmax1 = 0
xmax1 = n1

# hypothèse 2 : O2 est le réactif limitant
# n2 - 2*xmax2 = 0
xmax2 = n2 / 2

if xmax1 < xmax2:
    xf = xmax1
    print("Le méthane est le réactif limitant.")

elif xmax2 < xmax1:
    xf = xmax2
    print('Le dioxygène est le réactif limitant.')
    
elif xmax1 == xmax2:
    xf = xmax1
    print("Les réactifs ont été introduits dans des proportions stoechiométriques.")
    
# quantités de matière à l'état final   
n1f = n1 - xf    
n2f = n2 - 2*xf  
n3f = n3 + xf
n4f = n4 + 2*xf 
        
# affichage de l'état du système à l'état final
print("A l'état final, la quantité de matière du méthane vaut : ", n1f, " mol")
print("A l'état final, la quantité de matière de dioxygène vaut : ", n2f, " mol")
print("A l'état final, la quantité de matière dioxyde de carbone vaut : ", n3f, " mol")
print("A l'état final, la quantité de matière de l'eau vaut : ", n4f, " mol")
```

### Question 4 : réaction d’oxydation de l’éthanol par les ions dichromate

$3 C_{2}H_{5}OH_{~(aq)} + {2(Cr_{2}O_{7})^{2-}}_{(aq)} + {16H^{+}}_{(aq)} \rightarrow 3CH_{3}COOH_{~(aq)} + {4Cr^{3+}}_{(aq)} + {11H_{2}O}_{~(l)}$

```
# Bilan de matière de l'équation entre l'éthanol et les ions dichromate :
# 3C2H5OH + 2(Cr2O7)2- + 16H+ -> 3CH3COOH + 4Cr3+ + 11H2O
print("Bonjour, nous allons étudier la réaction entre l'éthanol et les ions dichromate")
m = float(input("Entrez la valeur de la masse d'étanol en g : "))
C1 = float(input("Entrez la valeur de la concentration de dichromate de potassium en mol/L : "))
V1 = float(input("Entrez la valeur du volume de dichromate de potassium en mL : "))

# quantité de matière initiale d'éthanol
M = 46 # masse molaire de l'éthanol en g/mol
n1 = m / M
print("La quantité de matière initiale en éthanol est de ", n1, "mol")

# quantité de matière initiale des ions dichromate
n2 = C1 * V1 * 0.001
print("La quantité de matière initiale en ion dichromate est de ", n2, "mol")

# quantité de matière des produits à l'état initial
n3 = 0
n4 = 0

# hypothèse 1 : l'éthanol est le réactif limitant
# n1 - 3*xmax1 = 0
xmax1 = n1 / 3

# hypothèse 2 : les ions dichromate sont le réactif limitant
# n2 - 2*xmax2 = 0
xmax2 = n2 / 2

if xmax1 < xmax2:
    xf = xmax1
    print("L'éthanol est le réactif limitant.")

elif xmax2 < xmax1:
    xf = xmax2
    print('Les ions dichromate sont le réactif limitant.')
    
elif xmax1 == xmax2:
    xf = xmax1
    print("Les réactifs ont été introduits dans des proportions stoechiométriques.")
    
# quantités de matière à l'état final   
n1f = n1 - 3*xf    
n2f = n2 - 2*xf  
n3f = n3 + 3*xf
n4f = n4 + 4*xf     
    
# affichage de l'état du système à l'état final
print("A l'état final, la quantité de matière de l'éthanol vaut : ", n1f, " mol")
print("A l'état final, la quantité de matière des ions dichromate vaut : ", n2f, " mol")
print("A l'état final, les ions H+ sont en excès")
print("A l'état final, la quantité de matière de l'acide éthanoique vaut : ", n3f, " mol")
print("A l'état final, la quantité de matière des ion chrome vaut : ", n4f, " mol")
print("A l'état final, l'eau est en excès")
```

### Question 5 : visualisation de l’avancement de la réaction de la production d’ammoniac dans le temps

```
# tracé du graphique de quantités de matière dans une réaction chimique

# importation des bibliothèques
# appel du module grpahique
from matplotlib import pyplot as plt
#appel du module numpy pour les calculs
import numpy as np

# Bilan de matière de l'équation : 3H2 + N2 -> 2NH3
print("Bonjour, nous allons étudier la réaction entre le dihydrogène H2 et le diazote N2")
n1 = float(input("Entrez la valeur de la quantité de matière de dihydrogène en mol : "))
n2 = float(input("Entrez la valeur de la quantité de matière de diazote en mol : "))
           
# quantité de matière du produit à l'état initial
n3 = 0

# hypothèse 1 : H2 est le réactif limitant
# n1 - 3*xmax1 = 0
xmax1 = n1 / 3

# hypothèse 2 : N2 est le réactif limitant
# n2 - xmax2 = 0
xmax2 = n2

if xmax1 < xmax2:
    xf = xmax1
    print("Le dihydrogène est le réactif limitant.")

elif xmax2 < xmax1:
    xf = xmax2
    print('Le diazote est le réactif limitant.')
    
elif xmax1 == xmax2:
    xf = xmax1
    print("Les réactifs ont été introduits dans des proportions stoechiométriques.")
    
# quantités de matière à l'état final   
n1f = n1 - 3*xf    
n2f = n2 - xf  
n3f = n3 + 2*xf      
           
# affichage de l'état du système à l'état final
print("A l'état final, la quantité de matière de dihydrogène vaut : ", n1f, " mol")
print("A l'état final, la quantité de matière de diazote vaut : ", n2f, " mol")
print("A l'état final, la quantité de matière d'ammoniac vaut : ", n3f, " mol")

# graphique
x = np.linspace(0, xf, 30)

plt.plot(x, n1 - 3*x, 'rX')  # courbe de la quantité de H2 en rouge
plt.plot(x, n2 - x, 'bX')    # courbe de la quantité de N2 en bleu
plt.plot(x, 2*x, 'go')       # courbe de la quantité de NH3 en vert

plt.xlabel("avancement en mol")
plt.ylabel("quantités de matière en mol")

labels = ["valeurs de n(H2)", 'valeurs de n(N2)', 'valeurs de n(NH3)']

plt.legend(labels) # affichage de la légende

plt.show()         # affichage du graphique
```
